### **1. Import Required Libraries**

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, Conv1D, MaxPooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, LearningRateScheduler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.sparse as sp
import joblib
from rdkit import Chem
from rdkit.Chem import MolToSmiles
```

### **2. Load and Preprocess Dataset**

```python
df = pd.read_csv("250k_rndm_zinc_drugs_clean_3.csv")  # Replace with your file path
print(df.head(1))
df['smiles'][1]
print("Shape - ",df.shape)
```

```python

def validate_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return mol is not None

df = df.dropna(subset=["smiles"])  # Drop missing SMILES
df = df[df["smiles"].apply(validate_smiles)]  # Keep only valid SMILES

def canonicalize_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return Chem.MolToSmiles(mol, canonical=True) if mol else None

df["canonical_smiles"] = df["smiles"].apply(canonicalize_smiles)

def tokenize_smiles(smiles):
    import re
    pattern = r"(\[[^\[\]]+\]|Br|Cl|Si|Se|Na|Li|K|B|C|N|O|F|P|S|I|H|c|n|o|p|s|b|r|\d+|=|#|\(|\)|\-|\+|\\|\/|@[A-Za-z])"
    return re.findall(pattern, smiles)

df["tokenized_smiles"] = df["canonical_smiles"].apply(tokenize_smiles)
```

### **3. SMILES Data Augmentation**

```python
def augment_smiles(smiles):
    """Generate an alternative SMILES representation"""
    mol = Chem.MolFromSmiles(smiles)
    return MolToSmiles(mol, doRandom=True) if mol else smiles  # Randomized SMILES

df["augmented_smiles"] = df["canonical_smiles"].apply(augment_smiles)
```

### **4. One-Hot Encoding**

```python
def one_hot_encode_smiles_sparse(df):
    unique_tokens = sorted(set(token for tokens in df['tokenized_smiles'] for token in tokens))
    token_to_index = {token: idx for idx, token in enumerate(unique_tokens)}
    vocab_size = len(unique_tokens)
    max_length = max(len(tokens) for tokens in df['tokenized_smiles'])
    one_hot_rows, one_hot_cols, one_hot_data = [], [], []
    for i, tokens in enumerate(df['tokenized_smiles']):
        for j, token in enumerate(tokens):
            token_idx = token_to_index[token]
            one_hot_rows.append(i)
            one_hot_cols.append(j * vocab_size + token_idx)
            one_hot_data.append(1)
    one_hot_matrix = sp.csr_matrix((one_hot_data, (one_hot_rows, one_hot_cols)), shape=(len(df), max_length * vocab_size), dtype=np.float32)
    return one_hot_matrix, token_to_index, vocab_size

one_hot_sparse, token_to_index, vocab_size = one_hot_encode_smiles_sparse(df)
```

### **5. Define Labels for Drug-Likeness**

```python
QED_THRESHOLD = 0.5  
df['drug_like'] = (df['qed'] >= QED_THRESHOLD).astype(int)
```

### **6. Split Dataset for Training**

```python
X_sparse = one_hot_sparse
Y = df['drug_like'].values
X_train, X_test, Y_train, Y_test = train_test_split(X_sparse, Y, test_size=0.2, random_state=42, stratify=Y)
max_length = max(len(tokens) for tokens in df['tokenized_smiles'])
```

### **7. Define and Train CNN-LSTM Model**

```python
def weighted_bce(alpha=0.75):
    def loss(y_true, y_pred):
        bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
        return tf.reduce_mean(alpha * y_true * bce + (1 - alpha) * (1 - y_true) * bce)
    return loss

def step_decay(epoch):
    initial_lr = 0.001
    drop = 0.5
    epochs_drop = 3
    lr = initial_lr * (drop ** (epoch // epochs_drop))
    return lr

model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(max_length, vocab_size)),
    MaxPooling1D(pool_size=2),
    Bidirectional(LSTM(128, return_sequences=True)),
    Dropout(0.3),
    LSTM(64, return_sequences=False),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(loss=weighted_bce(alpha=0.75), optimizer=Adam(learning_rate=0.0005), metrics=['accuracy'])

class_weight_dict = {0: 20.0, 1: 0.5}

checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_loss', mode='min')
early_stopping = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)
lr_scheduler = LearningRateScheduler(step_decay)

X_train = X_train.toarray().reshape(X_train.shape[0], max_length, vocab_size)
X_test = X_test.toarray().reshape(X_test.shape[0], max_length, vocab_size)

model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=5, batch_size=64, class_weight=class_weight_dict, callbacks=[checkpoint, early_stopping, lr_scheduler])
```

```python
X_test = X_test.reshape(X_test.shape[0], max_length, vocab_size)
```

### **8. Model Evaluation**

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(f"Test Accuracy: {accuracy:.4f}")
y_pred = (model.predict(X_test) > 0.5).astype(int)
print(classification_report(Y_test, y_pred, target_names=["Non-Drug-Like", "Drug-Like"]))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
cm = confusion_matrix(Y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Non-Drug-Like", "Drug-Like"], yticklabels=["Non-Drug-Like", "Drug-Like"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import joblib
from rdkit import Chem

# Load trained model and tokenizer
model_path = "best_model.keras"  # Load the new model
model = load_model(model_path)

tokenizer_path = "tokenizer.pkl"
token_to_index = joblib.load(tokenizer_path)

# Load other necessary files
vocab_size = len(token_to_index)
max_length = 71  # Ensure this matches the training configuration

def one_hot_encode_smiles(smiles):
    """One-hot encode a single SMILES string."""
    tokens = list(smiles)  # Adjust if needed
    one_hot_vector = np.zeros((max_length, vocab_size), dtype=np.float32)

    for j, token in enumerate(tokens[:max_length]):  # Limit sequence length
        if token in token_to_index:
            token_idx = token_to_index[token]
            one_hot_vector[j, token_idx] = 1

    return one_hot_vector.reshape(1, max_length, vocab_size)

def predict_drug_likeness():
    """Interactive function for predicting drug-likeness from SMILES input."""
    while True:
        smiles = input("Enter a chemical SMILES (or type 'exit' to quit): ")
        if smiles.lower() == 'exit':
            break

        # ✅ Try bypassing strict validation
        mol = Chem.MolFromSmiles(smiles, sanitize=False)  
        
        if mol is None:
            print("❌ Invalid SMILES string! Please enter a valid one.")
            continue

        # Preprocess SMILES
        encoded_input = one_hot_encode_smiles(smiles)

        # ✅ Run inference
        prediction = model.predict(encoded_input).flatten()[0]
        label = "Drug-Like ✅" if prediction >= 0.5 else "Non-Drug-Like ⚠️"

        print(f"\nInput SMILES: {smiles}")
        print(f"Prediction: {label} (Score: {prediction:.2f})\n")

# Run the prediction system
predict_drug_likeness()
```
